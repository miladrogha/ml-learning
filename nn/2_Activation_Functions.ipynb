```python
import matplotlib.pyplot as plt
import numpy as np
```

## Activation functions

### 1- Linear Activation Function:
- Use case: Output layer of a regression task.
- Equation: $ y = x $

<img src="./images/linear_activation_function.png" width=300 />


### 2- Sigmoid Activation Function
- Use case: Returns 0 for $-\infty $ to 0, 0.5 for 0, and 1 for $ >0 $
- Equation: $\frac{1}{1+ e^{-x}}$
- Pros: Adds non-linearity, the returned value contains info about the input, output is in the range of $[0,1]$

<img src="./images/sigmoid_activation_function.png" width=300 />

### 3- Rectified Linear Unit (ReLU) Activation Function
- Use case: Replaced Sigmoid. The most common activation function at the time.
- Equation: $ y = 0 \  \ \forall x \ \in [-\infty, 0) $ , 
            $ y = x \  \ \forall x \ \in [0, \infty) $
- Pros: Adds non-linearity, the returned value contains info about the input, output is in the range of $[0,1]$, is fast amd efficient.

<img src="./images/reLU_activation_function.png" width=300 />

```python
X = range(-10,10)
```

```python
y= X
y_sigmoid = [1/(1+ np.e **(-x)) for x in X]
plt.plot(X,y, c='teal', label="Linear Activation Function")
# plt.plot(X,y_sigmoid, c='r', label='Sigmoid Activation Function')
# plt.legend()
plt.savefig('./images/linear_activation_function.png')
```

```python
y_sigmoid = [1/(1+ np.e **(-x)) for x in X]
plt.plot(X,y_sigmoid, c='r', label='Sigmoid Activation Function')
plt.savefig('./images/sigmoid_activation_function.png')
```

```python
y_reLU = [x if x>0 else 0 for x in X ]
plt.plot(X,y_reLU, c='g', label='Sigmoid Activation Function')
plt.savefig('./images/reLU_activation_function.png')
```
